```python
import pandas as pd
url='https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/master/glassdoor_jobs.csv'
df = pd.read_csv(url,error_bad_lines=False)
```

```python
df
```

```python
# remove the glassdoor (salary estimate)
#remove the (\ndigits)
#arrange the revenue 

#remove the -1 of the salary 
df = df[df['Salary Estimate']!='-1']
# remove the (glassdoor  est.)
salary= df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary= salary[salary.apply(lambda x: 1 if 'per hour' in x.lower() else 0)!=1]
min_sd= salary.apply(lambda x: x.replace('$','').replace('K',''))

#min_sd.apply(lambda x: x.lower().replace('per hour',''))
```

```python
min_sd = min_sd.apply(lambda x: x.lower().replace('employer provided salary:',''))
df['min_salary']=min_sd.apply(lambda x: x.split('-')[0])
df['max_salary']=min_sd.apply(lambda x: x.split('-')[1])
```

```python
# apply to each row with axis=1 
df['Company Name_Revised']= df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-4],axis=1)

# job state 
# rate before 4
#df['job_state']= 
df['job_state']= df['Location'].apply(lambda x: x.split(',')[1])
df['job_state'].value_counts()
#CA, MA ,NY, VA 
```

```python
# check the headquarter 
df['place_Match_Head']= df.apply(lambda x: 1 if x.Location==x.Headquarters else 0,axis =1 )
df['place_Match_Head'].value_counts()
```
